```python
from ultralytics import YOLO
import glob
import os

# Load your trained model
model = YOLO("fire_model.pt")

# Path to your images folder
image_folder = "images"

# Get all image files (JPG, JPEG, PNG)
image_files = glob.glob(os.path.join(image_folder, "*.jpg")) + \
               glob.glob(os.path.join(image_folder, "*.jpeg")) + \
               glob.glob(os.path.join(image_folder, "*.png"))

# Create output folder (optional)
output_dir = "detected_results"
os.makedirs(output_dir, exist_ok=True)

# Run detection on each image
for img_path in image_files:
    results = model(img_path, save=True, project=output_dir, name="fire_detections")
    print(f"âœ… Detection done for: {img_path}")

print("ðŸ”¥ All detections completed. Check the 'detected_results/fire_detections' folder.")
```
